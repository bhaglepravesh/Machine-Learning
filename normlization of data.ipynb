```python
import numpy as np
def normlize_data(data):
    mean=np.mean(data)
    std=np.std(data)
    normalized_data=[(x-mean)/std for x in data]
    return normalized_data
```

```python
data = []
for i in range(0,100):
    data.append(np.random.randint(0,100))
print(data)
```

```python
print(normlize_data(data))
datas=normlize_data(data)
```

```python
print(np.mean(datas))
np.std(datas)
```

```python
def minmax_scaling(data):
    mins=min(data)
    maxs=max(data)
    norm_data=[(x-mins)/(maxs-mins) for x in data]
    return norm_data
```

```python
print(minmax_scaling(data))
```
